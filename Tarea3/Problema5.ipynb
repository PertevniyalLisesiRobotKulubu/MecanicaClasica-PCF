```python
#e > 1 -> Hyperbola

sin = np.sin
cos = np.cos
pi = np.pi

#Definitions

a = 1
e = 2

theta = np.linspace(-2,2, 360)

def r(par):
    return a/(1+e*cos(par))

plot = plt.polar(theta, r(theta), "red", label=r"$\displaystyle\epsilon > 1$")
plt.legend()
plt.show()
```

```python
#e = 1 -> Parabola

sin = np.sin
cos = np.cos
pi = np.pi

#Definitions

a = 1
e = 1

theta = np.linspace(-2,2, 360)

def r(par):
    return a/(1+e*cos(par))

plt.polar(theta, r(theta), "red",label=r"$\displaystyle\epsilon = 1$")
plt.legend()
plt.show()
```

```python
#e = 0 -> Circulo

sin = np.sin
cos = np.cos
pi = np.pi

#Definitions

a = 1
e = 0

theta = np.linspace(0,2*pi, 360)

def r(par):
    return a/(1+e*cos(par))

plt.polar(theta, r(theta), "red",label=r"$\displaystyle\epsilon = 0$")
plt.legend()
plt.show()
```

```python
# 0 < e < 1 -> Elipse

sin = np.sin
cos = np.cos
pi = np.pi

#Definitions

a = 1
e = 0.3

theta = np.linspace(0,2*pi, 360)

r1 = a/(1+0.45*cos(theta))
r2 = a/(1+0.65*cos(theta))
r3 = a/(1+0.8*cos(theta))


plt.polar(theta, r1, "red", label = r"$\displaystyle \epsilon = 0.45$")
plt.polar(theta, r2, "blue", label = r"$\displaystyle \epsilon = 0.6$")
plt.polar(theta, r3, "green", label = r"$\displaystyle \epsilon = 0.8$")
plt.legend()
plt.show()
```

```python
%%latex

Problema 5. $(\dot{\theta}$,r)
```

```python
r = np.linspace(1,10,100)
l = np.linspace(1,10,100)
m = np.linspace(1,10,100)

thetaDot = l/(m*r**2)

plt.plot(thetaDot,r,"red", label = r"$\displaystyle \dot{\theta} = l/mr^2$")
plt.legend()
plt.show()
```

```python
%%latex

Problema 5. $(\dot{r},\theta)$
```

```python
theta = np.linspace(-pi,pi, 360)
l = np.linspace(1,10,360)
m = np.linspace(1,10,360)
a = np.linspace(1,10,360)

sin = np.sin
cos = np.cos
pi = np.pi

rDot1 = (a*l*1*sin(theta))/m
rDot2 = (a*l*1.1*sin(theta))/m
rDot3 = (a*l*0.5*sin(theta))/m



plt.polar(theta,rDot1,"blue", label = r"$\displaystyle \epsilon = 1$")
plt.polar(theta,rDot2,"red", label = r"$\displaystyle \epsilon > 1$")
plt.polar(theta,rDot3,"green", label = r"$\displaystyle 0 < \epsilon < 1$")

plt.legend()
plt.show()
```

```python
%%latex

Problema 5. $(\dot{r},r)$
```

```python
Y, X = np.mgrid[-1.5:1.5:200j, 0.1:100:200j]

U = Y
V = 1 / X**2 - 1/X

plt.streamplot(X, Y, U, V,color=U,density=[1.3, 1.3], linewidth=2,
               cmap=plt.cm.winter)

plt.axhline(y=0, color="black")
plt.ylabel(r'$\dot{r}$', fontsize=30)
plt.xlabel(r'$r$',fontsize=30)

plt.show()
```
