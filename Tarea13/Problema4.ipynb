```python
from __future__ import print_function
import numpy as np
from numpy import pi,arctan
import matplotlib
matplotlib.use('nbagg')
import matplotlib.pyplot as plt

a=.1
b=.2
c=.3
d=.4
r = np.linspace(-20,20, 360)
theta =  (r*(a-1/r**2)**.5 *((a*r**2-1)**(.5)+arctan(1/((a*r**2-1)**.5))))/(a*r**2-1)**.5
theta1 = (r*(b-1/r**2)**.5 *((b*r**2-1)**(.5)+arctan(1/((b*r**2-1)**.5))))/(b*r**2-1)**.5
theta2 = (r*(c-1/r**2)**.5 *((c*r**2-1)**(.5)+arctan(1/((c*r**2-1)**.5))))/(c*r**2-1)**.5
theta3 = (r*(d-1/r**2)**.5 *((d*r**2-1)**(.5)+arctan(1/((d*r**2-1)**.5))))/(d*r**2-1)**.5

ax = plt.subplot(111, projection='polar')
plt.polar(theta, r, color='r', linewidth=3)
plt.polar(theta1, r, color='b', linewidth=3)
plt.polar(theta2, r, color='g', linewidth=3)
plt.polar(theta3, r, color='black', linewidth=3)

ax.grid(True)

plt.show()
```
