```python
#Tarea 2. Problema 1.

from __future__ import print_function
import numpy as np
import matplotlib
matplotlib.use('nbagg')
import matplotlib.pyplot as plt
from numpy import *
from pylab import *

# LaTeX format
matplotlib.rc('font', family='serif')
matplotlib.rc('font', size=16)
matplotlib.rc('legend', fontsize=16)
matplotlib.rc('legend', numpoints=1)
matplotlib.rc('legend', handlelength=1.5)
matplotlib.rc('legend', frameon=False)
matplotlib.rc('xtick.major', pad=7)
matplotlib.rc('xtick.minor', pad=7)
matplotlib.rc('text', usetex=True)
matplotlib.rc('text.latex', 
              preamble=[r'\usepackage[T1]{fontenc}',
                        r'\usepackage{amsmath}',
                        r'\usepackage{txfonts}',
                        r'\usepackage{textcomp}'])
close('all')
```

```python
#Streamplot parte 1
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
a = np.mgrid[-10:10:100j]
U = X - Y - X*(X**2+Y**2)
V = X + Y - Y*(X**2+Y**2)

plt.streamplot(X, Y, U, V,color=V,density=[1.2, 1.2], linewidth=2,
               cmap=plt.cm.winter)

plt.show()
```

```python
#Stream plot parte b. Centros

Y, X = np.mgrid[-1.5:1.5:100j, -1.5:1.5:100j]

U = Y
V = X - (X**3)

plt.streamplot(X, Y, U, V,color=U,density=[1.4, 1.4], linewidth=2,
               cmap=plt.cm.autumn)

plt.show()
```

```python
#Stream plot parte b. Focos inestables

Y, X = np.mgrid[-1.5:1.5:100j, -1.5:1.5:100j]

U = Y
V = Y + X - (X**3)

plt.streamplot(X, Y, U, V,color=U,density=[1.4, 1.4], linewidth=2,
               cmap=plt.cm.viridis)

plt.show()
```

```python
#Stream plot parte b. Focos estables

Y, X = np.mgrid[-1.5:1.5:100j, -1.5:1.5:100j]

U = Y
V = -Y + X - (X**3)

plt.streamplot(X, Y, U, V,color=U,density=[1.4, 1.4], linewidth=2,
               cmap=plt.cm.plasma)

plt.show()
```

```python
#Stream plot parte b. Nodos Estables

Y, X = np.mgrid[-3:3:500j, -3:3:500j]

U = Y
V = -10*Y + X - (X**3)

plt.streamplot(X, Y, U, V,color=U,density=[1.4, 1.4], linewidth=2,
               cmap=plt.cm.winter)

plt.show()
```

```python
#Stream plot parte b. Nodos Inestable

Y, X = np.mgrid[-3:3:500j, -3:3:500j]

U = Y
V = 10*Y + X - (X**3)

plt.streamplot(X, Y, U, V,color=U,density=[1.4, 1.4], linewidth=2,
               cmap=plt.cm.summer)

plt.show()
```
