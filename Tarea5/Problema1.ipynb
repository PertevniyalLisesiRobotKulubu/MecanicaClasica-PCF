```python
from __future__ import print_function
import numpy as np
from numpy import arccos,sin, cos, pi, array, sqrt
import matplotlib
matplotlib.use('nbagg')
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)

#LaTeX
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
#Effective potential graph

#Equal masses

#Parameters
l = 1
M = 1
m = 1
px = 1
theta = np.linspace(-pi,pi,360)
g = 9.8

#Effective potential def.
V_ef = (px**2/(m+M)) - m*g*l*cos(theta)

#Energies
plt.axhline(y=-9.3, color="blue")
plt.axhline(y=5, color="green")
plt.axhline(y=-3, color="orange")

#x-axis
plt.axhline(y=0, color="black")

#Plot and style
plt.plot(theta,V_ef,"red")
plt.xticks([])
plt.yticks([])

# Text in plot
plt.text(3,6, r"$E_1$", bbox=dict(facecolor='white', alpha=0.1),fontsize=16)
plt.text(3, -8.4, r"$E_2$", bbox=dict(facecolor='white', alpha=0.1),fontsize=16)
plt.text(3, -2, r"$E_3$", bbox=dict(facecolor='white', alpha=0.1),fontsize=16)

plt.text(-1, 9, r"Masas iguales", bbox=dict(facecolor='white', alpha=0.1),fontsize=16)


plt.text(4.25, 0.7, r"$\theta = 0$", bbox=dict(facecolor='white', alpha=0.1),fontsize=16)
plt.ylabel(r'$V_{ef}$', fontsize=30)
plt.xlabel(r'$\theta$',fontsize=30)

plt.show()
```
